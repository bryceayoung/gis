## Using Sequences Effectively

````python
# Define a function that calculates haversine distance between two places

import numpy as np

def haversine(p1, p2):
    x1, y1 = map(np.deg2rad, p1)
    x2, y2 = map(np.deg2rad, p2)
    dphi = np.abs(y2 - y1) # Difference in latitude
    dlambda = np.abs(x2 - x1) # Difference in longitude
    angle = 2 * np.arcsin(np.sqrt(np.add(
        np.power(np.sin(dphi / 2), 2),
        np.cos(y1) * np.cos(y2) * np.power(np.sin(dlambda / 2), 2)
    )))
    return float(angle * 6371e3) # Earth's radius: 6371e3 meters
````

````python
# Longitude, Latitude
MISSOULA = (-114.011667, 46.8625)

# Distance between Missoula and Brainerd, MN?
haversine(MISSOULA, (-94.2, 46.4))
````

````python
# Find distance of a range of cities to Missoula

lng_list = places['longitude'].tolist()
lat_list = places['latitude'].tolist()

for i in range(0, 10):
    lng = lng_list[i]
    lat = lat_list[i]
    dist = haversine(missoula, [lng, lat])
    print(cities[i], ':', f'{(dist / 1000):.2f}', 'km')
````

### Challenge: Closest City to Missoula?

**What's the closest city in this dataset to Missoula?** Remember that there are built in `min()` and `max()` functions, and you can call the `index()` method on a `list` to get the numeric position of an element.

````python
# Find closest city to Missoula in entire range of cities

distances = []

for i in range(0, 768):
    lng = lng_list[i]
    lat = lat_list[i]
    if cities[i] != 'Missoula':
        dist = haversine(missoula, [lng, lat])
    else:
        dist = 1e9
    distances.append(dist)

min_index = distances.index(min(distances))

print('Closest city:', cities[min_index], f'{(min(distances)/1000):.2f}', 'km')
````
