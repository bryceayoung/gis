## Problem 3.3

**How many kilometers would you have to travel to visit every "Fish Creek" in Montana?** Don't worry about whether the route is the shortest possible route, just calculate the distance along *any* route that will bring you to *every* Fish Creek, even if you're visiting them in a random, zig-zagging order. You'll need to use the whole `geo` table for this...

````python
geo.columns
````

**The idea is to start at one Fish Creek (any one) and travel to the next, then the next, and so on.** The order in which you visit Fish Creeks doesn't matter.

Also, remember the `haversine()` function:

````python
import numpy as np

def haversine(p1, p2):
    x1, y1 = map(np.deg2rad, p1)
    x2, y2 = map(np.deg2rad, p2)
    dphi = np.abs(y2 - y1) # Difference in latitude
    dlambda = np.abs(x2 - x1) # Difference in longitude
    angle = 2 * np.arcsin(np.sqrt(np.add(
        np.power(np.sin(dphi / 2), 2),
        np.cos(y1) * np.cos(y2) * np.power(np.sin(dlambda / 2), 2)
    )))
    return float(angle * 6371e3) # Earth's radius: 6371e3 meters
````

**And remember that `haversine()` takes two sequences as arguments;**

```py
p1 = (some_longitude, some_latitude)
p2 = (another_longitude, another_latitude)
```

#### ANSWER 3.3

This code block (below) finds distance between points if travelling from place to place based on the order of the list 'coords'

````python
# Create a df where 'name' == 'Fish Creek'
# This ensures that subsequent lists will be indexed the same
fishcreek_df = geo[geo['name'] == 'Fish Creek']

# Zip lat and long and save as list
lng = fishcreek_df['longitude'].to_list()
lat = fishcreek_df['latitude'].to_list()

coords = list(zip(lng, lat))

# Calculate the total distance that connects all the fish creeks.
# This starts at the top of the list and goes through the bottom.

def get_distances(coords):
    '''
    Calculate and store the distances between consecutive points as a list of distances (km).
    
    Inputs
    --------------
    coords = a list of tuples of longitude and latitude (e.g. coords = list(zip(lng, lat)))
    longitude and latitude must be in decimal degrees
    haversine function must be defined as 'haversine'
    
    Outputs
    -------------
    Kilometers rounded to 2 decimal places
    
    '''
    # Initiate a list called 'distances'
    distances = []

    # Starting at 1, calculate distance to 2, store as 'dist', then 2 to 3 etc.
    for i, place in enumerate(coords):
        if (i + 1) >= len(coords):                                     # Repeat for len(coords) minus 1
            break
        else:
            dist = round(haversine(coords[i], coords[i+1])/1000, 2)    # p1 = coords[i], p2 = coords[i + 1]
            distances.append(dist)
    return distances

    for i in range(0, len(coords)-1):
        dist = round(haversine(coords[i], coords[i+1])/1000, 2)

# Sum values of 'distances'
distances = get_distances(coords)
total = sum(distances)

print(f'The total distance connecting all Fish Creeks is {total} km')
````

This code block (below) starts at the first element of 'coords' and subsequently chooses the next coordinate based on the smallest haversine distance. The result 'total' is the distance traveled if moving from one coordinate to the next closest coordinate until all coordinates have been visited.

Caveat: **I had to ask ChatGPT for the `while` loop** but thought I would include the code anyways since it was a good problem to solve.

Other caveat: This assumes that the first location in 'coords' is the most efficient location to start, which it is not necessarily.

````python
import numpy as np
import pandas as pd

###-----NOTE-----###
# The 'haversine' and 'get_distances' functions used here are defined in previous code blocks
###--------------###

# Create a df where geo['name'] == 'Fish Creek'
# This ensures that subsequent lists will be indexed the same
fishcreek_df = geo[geo['name'] == 'Fish Creek']

# Zip lat and long and save as list
lng = fishcreek_df['longitude'].to_list()
lat = fishcreek_df['latitude'].to_list()

coords = list(zip(lng, lat))

# Define function to minimize distance between consecutive points
def reorder_coordinates(coords):
   
    """
    Reorder coordinates to minimize the total Haversine distance between consecutive points.
    """

    reordered = [coords.pop(0)]       # Start with the first coordinate, removing it from 'coords' and 
                                      # placing in new list 'reordered'
    
    while coords:                     # Executes as long as there are still elements in 'coords'
        last = reordered[-1]          # Accesses the last element in 'reordered'
        next_coord = min(coords, key = lambda x: haversine(last, x)) # Use haversine distance to define how `min()` calculates min
        reordered.append(next_coord)  
        coords.remove(next_coord)     
    
    return reordered

# Reorder coordinates to minimize distance
reordered_coords = reorder_coordinates(coords)

# Obtain distances between reordered coordinates
min_distances = get_distances(reordered_coords)

# Calculate the total distance
total = sum(min_distances)
print(f"The minimum distance connecting all Fish Creeks is {total} km")
````

#### ANSWER 3.4

````python
# Reduce the df so that 'flag' == 1 or 5
df = co2[(co2['flag'] == 1) | (co2['flag'] == 5)].copy()

# Define 'dms2dd' function to convert deg, min, sec to decimal deg
def dms2dd(degrees, minutes, seconds = 0):
    return round(degrees + (minutes / 60) + (seconds / (60 * 60)), 5)

# Create an empty list for 'latitude' and 'longitude'
lat = []
lng = []

# Use dms2dd to create two new lists for lat and long in decimal degrees
for row in df.to_records():
    dd = dms2dd(row['lat_degrees'], row['lat_minutes'])
    if 'S' in row['lat_ns']:
        dd = -dd                                       # Make dd negative if coord in southern hemisphere
    lat.append(dd)
    dd = dms2dd(row['lng_degrees'], row['lng_minutes'])
    if 'W' in row['lng_ew']:
        dd = -dd                                       # Make dd negative if coord in western hemisphere
    lng.append(dd)

# Convert the df to a list of rows containing the columns 'flight' 'date' 'time' 'lat_dd' 'lng_dd' 'pressure_hPa' and 'co2_ppm'
# Add dd columns to df
df.loc[:, 'lat_dd'] = lat
df.loc[:, 'lng_dd'] = lng

# Reduce dimensionality and save as new df
df1 = df[['flight', 'date', 'time', 'lat_dd', 'lng_dd', 'pressure_hPa', 'co2_ppm']]

# Initiate empty list for rows
df1_rows = []

# Place rows in list
for row in df1.to_records():
    df1_rows.append(row)
    
df1_rows[0:5]
````
